```python
import tkinter as tk
import math
def button_click(value):
    current=entry.get()
    if value=="=":
        result=eval(current)
        entry.delete(0,tk.END)
        entry.insert(tk.END,result)
    elif value=="C":
        entry.delete(0,tk.END) 
    elif value=="X":
        entry.delete(0)
    else:
        entry.insert(tk.END,value)
        
# Create the main window
root = tk.Tk()
root.title("Calculator")
entry=tk.Entry(width="40",bg="pink")
entry.grid()
#subtraction
button_subtract=tk.Button(root,text="-",command=lambda:button_click("-"),bg="Yellow",padx="2",pady="3")
button_subtract.grid(row=0,column=2)
#addtion
button_addition=tk.Button(root,text="+",command=lambda:button_click("+"),bg="sky blue",padx="2",pady="3")
button_addition.grid(row=0,column=3)
#mul
button_mul=tk.Button(root,text="*",command=lambda:button_click("*"),bg="Pink",padx="2",pady="3")
button_mul.grid(row=0,column=4)
#division
button_div=tk.Button(root,text="/",command=lambda:button_click("/"),bg="Orange",padx="2",pady="3")
button_div.grid(row=5,column=4)
#sin()
button_sin=tk.Button(root,text="sin",command=lambda:button_click("math.sin"),padx="2",pady="3")
button_sin.grid(row=5,column=2)
#cos()
button_cos=tk.Button(root,text="cos",command=lambda:button_click("math.cos"),padx="2",pady="3")
button_cos.grid(row=1,column=4)    
#log()
button_log=tk.Button(root,text="log",command=lambda:button_click("math.log"),padx="2",pady="3")
button_log.grid(row=2,column=4)
#sqrt()
button_sqrt=tk.Button(root,text="sqrt",command=lambda:button_click("math.sqrt"),padx="2",pady="3")
button_sqrt.grid(row=3,column=4)
#equals to
button_equals=tk.Button(root,text="=",command=lambda:button_click("="),padx="2",pady="3")
button_equals.grid(row=4,column=1)
# C button
button_C=tk.Button(root,text="C",command=lambda:button_click("C"),bg="Sky blue",padx="2",pady="3")
button_C.grid(row=4,column=4)
#7
button_7=tk.Button(root,text="7",command=lambda:button_click("7"),padx="2",pady="3")
button_7.grid(row=1,column=1)
#8
button_8=tk.Button(root,text="8",command=lambda:button_click("8"),padx="2",pady="3")
button_8.grid(row=1,column=2)
#9
button_9=tk.Button(root,text="9",command=lambda:button_click("9"),padx="2",pady="3")
button_9.grid(row=1,column=3)
#4
button_4=tk.Button(root,text="4",command=lambda:button_click("4"),padx="2",pady="3")
button_4.grid(row=2,column=1)
#5
button_5=tk.Button(root,text="5",command=lambda:button_click("5"),padx="2",pady="3")
button_5.grid(row=2,column=2)
#6
button_6=tk.Button(root,text="6",command=lambda:button_click("6"),padx="2",pady="3")
button_6.grid(row=2,column=3)
#1
button_1=tk.Button(root,text="1",command=lambda:button_click("1"),padx="2",pady="3")
button_1.grid(row=3,column=1)
#2
button_2=tk.Button(root,text="2",command=lambda:button_click("2"),padx="2",pady="3")
button_2.grid(row=3,column=2)
#3
button_3=tk.Button(root,text="3",command=lambda:button_click("3"),padx="2",pady="3")
button_3.grid(row=3,column=3)
#0
button_0=tk.Button(root,text="0",command=lambda:button_click("0"),padx="2",pady="3")
button_0.grid(row=4,column=2)
#00
button_00=tk.Button(root,text="00",command=lambda:button_click("00"),padx="2",pady="3")
button_00.grid(row=4,column=3)
#X
button_X=tk.Button(root,text="X",command=lambda:button_click("X"),bg="Red",padx="2",pady="3")
button_X.grid(row=0,column=1)
#(
button_h=tk.Button(root,text="( )",command=lambda:button_click("( )"),padx="2",pady="3")
button_h.grid(row=5,column=3)
#.
button_g=tk.Button(root,text=".",command=lambda:button_click("."),padx="3",pady="3")
button_g.grid(row=5,column=1)
# Run the Tkinter event loop
root.mainloop()  



```
