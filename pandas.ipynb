```python
import pandas as pd
# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve','Alice'],
        'Department': ['HR', 'IT', 'HR', 'IT', 'HR','HR'],
        'Salary': [5000, 6000, 5500, 7000, 6000,5000]}
df = pd.DataFrame(data)
# Group by 'Department' and calculate the average salary for each department
grouped = df.groupby('Department')['Salary'].mean() 
print(grouped)
```

```python
import pandas as pd
# Sample DataFrame
data = {'col1': [10, 20, 30, 40], 'col2': [50, 60, 70, 80], 'col3': [90, 100, 110, 120]}
df = pd.DataFrame(data)
# Select a specific row (third row)
print(df.iloc[2])  # Row at index 2 (third row)
# Select a specific column (second column)
print(df.iloc[:, 1])  # Selects all rows from column 1 (second column)
# Select specific rows and columns
print(df.iloc[1:3, 0:2])  # Rows 1 and 2, Columns 0 and 1
# Select a single value (second row, third column)
print(df.iloc[1, 2])  # Value at position (1, 2), which is the second row, third column
# Conditional slicing (selecting all rows and specific columns based on index positions)
print(df.iloc[1:3, [0, 2]])  # Selects rows 1-2 and columns 0 and 2
df.sort_values(by='col1')  # Sorts the DataFrame by a specific column
```

```python
# find correlation 
import pandas as pd 
data={
    'A':[1,2,3,4,5],
    'B':[5,4,3,2,1],
    'C':[1,3,5,7,9]
}
df=pd.DataFrame(data)
#find correlation
matrix=df.corr()
print(matrix)
```

```python
# Find correlation between two library
import pandas as pd 
data={
    'A':[1,2,3,4,5],
    'B':[5,4,3,2,1],
}
df=pd.DataFrame(data)
#find correlation between column A and column B
correlation=df['A'].corr(df['B'])
print(f"correlation between A and B:{correlation}") 
```
