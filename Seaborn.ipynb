```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# importing datasets
#print(sns.get_dataset_names())  # all the datasets that are available 
df = sns.load_dataset("tips")    # load_dataset is used to load the dataset that is avialable in ourÂ sns
print(df)
```

```python
sns.displot(df['tip'],kde=False)  # kde flag is set to false as a result the representation of kernal estimate plot will be removed and only histogram is plotted. 
plt.show()       # kernal density estimation - is the way to estimate the probability density function of continue random variable
#used for non parametricÂ analysis
```

```python
sns.displot(df['tip'],kind='hist')
plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
plt.show()
# hebinplotÂ =Â hex
#Â 
#
```

```python
sns.barplot(x='tip',y='total_bill',data=df)
plt.show()
```

```python
sns.countplot(x='tip',data=df)
plt.show()
```

```python
sns.pointplot(x='tip',y='total_bill',data=df)
plt.show()
```

cr=df[['total_bill','tip']].corr() # Correlation
sns.heatmap(cr,annot=True,cmap='coolwarm')
plt.show() 

```python
sns.pairplot(df,hue='sex')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
#lineplot
sns.lineplot(x='total_bill',y='tip',data=df)
```

```python
#regplot
sns.regplot(x='tip',y='size',data=df)
```

```python
#lmplot
sns.lmplot(x='total_bill',y='tip',data=df,hue='size')
```

```python
#factor plot
sns.catplot(x='total_bill',y='smoker',data=df,kind='bar')
```

```python
#catplot
sns.catplot(x='total_bill',y='tip',data=df,kind='box')
```
