```python
#TASK 1
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\rk871\Desktop\First Task.csv")
print(df) 
#(s.to_string())
```

```python
df=df.drop_duplicates()
df
```

```python
df_cleaned=df.dropna()
df_cleaned
```

```python
df = df.dropna(subset=['Product_name'])
df
```

```python
# Calculation sum
total_quantity = df['Quantity'].sum()
print(f"Total order quantity: {total_quantity} kg") 
```

```python
#Mean
total_quantity=df['Quantity'].mean()
print(f"Total order quantity:{total_quantity}kg")
```

```python
#Median
total_quantity=df['Quantity'].median()
print(f"Total order quantity:{total_quantity}kg")
```

```python
#Max
total_quantity=df['Quantity'].max()
print(f"Total order quantity:{total_quantity}kg")
```

```python
#Min
total_quantity=df['Quantity'].min()
print(f"Total order quantity:{total_quantity}kg")
```

```python
#Sales Trend over time
plt.figure(figsize=(20,10))
plt.bar(df["Product_name"],df["Quantity"])
plt.xlabel("product")
plt.ylabel("sales")
plt.title("Data analysis ")
plt.show()
```
